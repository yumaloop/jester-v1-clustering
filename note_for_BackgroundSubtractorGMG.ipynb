```python
import os
import cv2
import numpy as np
from PIL import Image

from copy import deepcopy

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_prv_gray = cv2.imread('data/video/20bn-jester-v1/100041/00001.jpg', cv2.IMREAD_GRAYSCALE)
img_nxt_gray = cv2.imread('data/video/20bn-jester-v1/100041/00020.jpg', cv2.IMREAD_GRAYSCALE)
img_prv_gray = cv2.resize(img_prv_gray, (48,48))
img_nxt_gray = cv2.resize(img_nxt_gray, (48,48))

print("img_prv_gray.shape :", img_prv_gray.shape)
print("img_nxt_gray.shape :", img_nxt_gray.shape)

plt.figure()
plt.imshow( Image.fromarray(np.uint8(img_prv_gray)) )
plt.gray()

plt.figure()
plt.imshow( Image.fromarray(np.uint8(img_nxt_gray)) )
plt.gray()
```

```python
frame = img_nxt_gray
plt.figure()
plt.imshow( Image.fromarray(np.uint8(frame)) )
plt.gray()
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
fgbg = cv2.bgsegm.createBackgroundSubtractorGMG()
```

```python
fgmask = fgbg.apply(frame)
fgmask = cv2.morphologyEx(fgmask, cv2.MORPH_OPEN, kernel)

plt.figure()
plt.imshow( Image.fromarray(np.uint8(fgmask)) )
plt.gray()
```

```python
import os
import cv2
import numpy as np
from PIL import Image
from copy import deepcopy
import matplotlib.pyplot as plt
%matplotlib inline

img_src1 = cv2.imread('data/video/20bn-jester-v1/100041/00001.jpg')
img_src2 = cv2.imread('data/video/20bn-jester-v1/100041/00002.jpg')
img_src1 = cv2.resize(img_src1, (48,48))
img_src2 = cv2.resize(img_src2, (48,48))

fgbg = cv2.bgsegm.createBackgroundSubtractorMOG()

fgmask = fgbg.apply(img_src1)
fgmask = fgbg.apply(img_src2)

plt.imshow(fgmask)
```
